```python
# Created on May 7, 2018
# Author: Eun Sun Lee

#with open('valencia_aquamote4_nurseshark_woheader.csv', 'rb') as csvfile:
from numpy import genfromtxt
my_data = genfromtxt('valencia_aquamote4_nurseshark_woheader.csv', delimiter=',')
```

```python
my_data[0][0] = 0.636230469
print(my_data[0])
```

```python
import numpy as np
import math
time_list = []
for j in my_data: 
    time_list.append(j[11])
    
for i in range(0,len(time_list)+1,20):
    if i == 0:
        pass
    elif i == 20: 
        x = np.linspace(40, time_list[i-1], 21)
        time_list[0:20] = x[1:21]
    else:
        x = np.linspace(time_list[i-21], time_list[i-1], 21)
        time_list[i-20:i] = x[1:21]
```

```python
#Accelerometer 
accelx_list = []
accely_list = []
accelz_list = []
#Gyroscope
gyrox_list = []
gyroy_list = []
gyroz_list = []
#Magnetometer
magx_list = []
magy_list = []
magz_list = []
#Pressure and Temperature 
pressure_list = []
temp_list = []

for i in my_data: 
    accelx_list.append(i[0])
    accely_list.append(i[1])
    accelz_list.append(i[2])
    
    gyrox_list.append(i[3])
    gyroy_list.append(i[4])
    gyroz_list.append(i[5])

    magx_list.append(i[6])
    magy_list.append(i[7])
    magz_list.append(i[8])
    
    pressure_list.append(i[9])
    temp_list.append(i[10])
```

```python
import matplotlib.pyplot as plt
plt.figure(1, figsize = (6,6))
plt.subplot(311)
plt.plot(time_list, accelx_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Accel x-axis')
plt.subplot(312)
plt.plot(time_list, accely_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Accel y-axis')
plt.subplot(313)
plt.plot(time_list, accelz_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Accel z-axis')
plt.show()

plt.figure(2, figsize = (6,6))
plt.subplot(311)
plt.plot(time_list, gyrox_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Gyro x-axis')
plt.subplot(312)
plt.plot(time_list, gyroy_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Gyro y-axis')
plt.subplot(313)
plt.plot(time_list, gyroz_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Gyro z-axis')
plt.show()

plt.figure(3, figsize = (6,6))
plt.subplot(311)
plt.plot(time_list, magx_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Mag x-axis')
plt.subplot(312)
plt.plot(time_list, magy_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Mag y-axis')
plt.subplot(313)
plt.plot(time_list, magz_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Mag z-axis')
plt.show()



plt.figure(4)
plt.subplot(211)
plt.plot(time_list, temp_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Temperature')

plt.subplot(212)
plt.plot(time_list, pressure_list, '.')
plt.xlabel('time(s)')
plt.ylabel('Pressure')
plt.show()
```

```python
def zoominfunc(x,y):
    time = time_list[x:y]
    accelx = accelx_list[x:y]
    accely = accely_list[x:y]
    accelz = accelz_list[x:y]
    gyrox = gyrox_list[x:y]
    gyroy = gyroy_list[x:y]
    gyroz = gyroz_list[x:y]
    magx = magx_list[x:y]
    magy = magy_list[x:y]
    magz = magz_list[x:y]
    pressure = pressure_list[x:y]
    temp = temp_list[x:y]



    plt.figure(1,figsize = (6,6))
    plt.subplot(311)
    plt.plot(time, accelx, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Accel x-axis')
    plt.subplot(312)
    plt.plot(time, accely, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Accel y-axis')
    plt.subplot(313)
    plt.plot(time, accelz, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Accel z-axis')
    plt.show()

    plt.figure(2, figsize = (6,6))
    plt.subplot(311)
    plt.plot(time, gyrox, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Gyro x-axis')
    plt.subplot(312)
    plt.plot(time, gyroy, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Gyro y-axis')
    plt.subplot(313)
    plt.plot(time, gyroz, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Gyro z-axis')
    plt.show()

    plt.figure(3, figsize = (6,6))
    plt.subplot(311)
    plt.plot(time, magx, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Mag x-axis')
    plt.subplot(312)
    plt.plot(time, magy, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Mag y-axis')
    plt.subplot(313)
    plt.plot(time, magz, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Mag z-axis')
    plt.show()

    plt.figure(4)
    plt.subplot(211)
    plt.plot(time, temp, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Temperature')

    plt.subplot(212)
    plt.plot(time, pressure, '.')
    plt.xlabel('time(s)')
    plt.ylabel('Pressure')
    plt.show()
```
