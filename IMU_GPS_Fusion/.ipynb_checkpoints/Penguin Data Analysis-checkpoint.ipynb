```python
print(len(data))
short_len = int(len(data))
```

## Read GPS data

## Visualizing GPS Data

```python

lat = []
long = []
alt = []
latitude = 0
longitude = 0
altitude = 0 

term = []
val = 0
for i in range(len(Latitude)):
    if i == 0:
        pass
    else:
        if (math.isnan(Latitude[i])):
            val = val + 1
            #pass
        else:
            term.append(val)
            val = 0
            latitude = Latitude[i]
            longitude = Longitude[i]
            altitude = Altitude[i]
        lat.append(latitude)
        long.append(longitude)
        alt.append(altitude)
        
```

## How often the penguin came out of the water

```python
# print(term)
plt.plot(term)
```

```python
x = np.array(term)
y = np.bincount(x)
plt.plot(y)
```

## 2D Plot

```python
fig = plt.figure(figsize=(20,10))
plt.plot(lat, long)
plt.show()
```

## 3D Plot

```python


mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.plot(lat, long, alt, label='GPS trace')
ax.legend()

plt.show()
```

```python
short_data = data.head(n=short_len)#cv 
Acc_x = short_data['Acc_x'].tolist()
Acc_y = short_data['Acc_y'].tolist()
Acc_z = short_data['Acc_z'].tolist()
```

```python
print("Minimum Acceleration: ", min(abs(np.array(Acc_x))))
print("Maximum Acceleration: ", max(abs(np.array(Acc_x))))
print("Mean Acceleration: ", np.mean(abs(np.array(Acc_x))))

print("Minimum Acceleration: ", min(abs(np.array(Acc_y))))
print("Maximum Acceleration: ", max(abs(np.array(Acc_y))))
print("Mean Acceleration: ", np.mean(abs(np.array(Acc_y))))

print("Minimum Acceleration: ", min(abs(np.array(Acc_z))))
print("Maximum Acceleration: ", max(abs(np.array(Acc_z))))
print("Mean Acceleration: ", np.mean(abs(np.array(Acc_z))))

# Accleration includes gravity so possible that one axis 
# such as z was more affected by the gravity depends on the way the sensor was mounted.
```
