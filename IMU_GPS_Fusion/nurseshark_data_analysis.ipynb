```python
import math
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
data = pd.read_csv('20180524-192530-MTCC1_all.csv',  encoding = "ISO-8859-1")
# header = data.columns.values
# print(header)
```

```python
header = data.columns.values
print(header)
print(len(data))
short_len = int(len(data))
print(short_len)
```

```python
short_data = data.head(n=short_len)#cv 
ax = short_data['Accelerometer X [m/s²]'].tolist()
ay = short_data['Accelerometer Y [m/s²]'].tolist()
az = short_data['Accelerometer Z [m/s²]'].tolist()
gx = short_data['Gyroscope X [mrad/s]'].tolist()
gy = short_data['Gyroscope Y [mrad/s]'].tolist()
gz = short_data['Gyroscope Z [mrad/s]'].tolist()
```

```python
plt.title("Acceleration")
plt.plot(ax)
plt.plot(ay)
plt.plot(az)
plt.show()

plt.title("Gyroscope")
plt.plot(gx)
plt.plot(gy)
plt.plot(gz)
plt.show()

```

```python
# a_ = np.array([3,-1,-6,4,7,2,9,9,9])
# # n = np.bincount(a-np.min(a))
# # no = n[n>0]
# # ao = np.unique(np.sort(a))
# # plt.plot(ao, no)
import math

def freqgraph(x):
    x_10 = np.array(x)*10
    x_f10 = np.floor(x_10)
    x_int = x_f10.astype(int)
    x_sort = np.unique(np.sort(x_int))/10
    n = np.bincount(x_int-np.min(x_int))
    no = n[n>0]
    plt.plot(x_sort,no)
    plt.show()

freqgraph(ax)
freqgraph(ay)
freqgraph(az)
```
