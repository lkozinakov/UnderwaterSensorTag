```python
import wireframe as wf
import wireframeDisplay as wd
import pygame
import math
import numpy as np
import csv


def IMUtoEulersangle(IMUdata):
    # similar to: https://gist.github.com/shoebahmedadeel/0d8ca4eaa65664492cf1db2ab3a9e572

    accelx, accely, accelz = IMUdata[0], IMUdata[1], IMUdata[2]
    gyrox, gyroy, gyroz = IMUdata[3], IMUdata[4], IMUdata[5]
    magx, magy, magz = IMUdata[6], IMUdata[7], IMUdata[8]
    
    pitch = np.arctan2(accely, np.sqrt((accelx*accelx)+(accelz*accelz)))
    roll = np.arctan2(-accelx, np.sqrt((accely*accely)+(accelz*accelz)))

    Yh = (magy * math.cos(roll)) - (magz * math.sin(roll))
    Xh = (magx * math.cos(pitch))+(magy * math.sin(roll)*math.sin(pitch)) + (magz * math.cos(roll) * math.sin(pitch))
    yaw =  np.arctan2(Yh, Xh)

    roll = roll
    pitch = pitch
    yaw = yaw
    
    return roll, pitch, yaw


k = 0
time = []
accelX = []
accelY = []
accelZ = []
gyroX = []
gyroY = []
gyroZ = []
magX = []
magY = []
magZ = []

with open('IMU_test_straight_square_circle_20180402_191102.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        if k> 15000 and k < 15600:
            accelX.append(row[0])
            accelY.append(row[1])
            accelZ.append(row[2])
            gyroX.append(row[3])
            gyroY.append(row[3])
            gyroZ.append(row[3])
            magX.append(row[3])
            magY.append(row[3])
            magZ.append(row[3])
        k = k + 1 
    
        
fs = 40
time = np.linspace(0,len(accelX)/fs, len(accelX))

time = np.array(time).astype(np.float)
accelX = np.array(accelX).astype(np.float)
accelY = np.array(accelY).astype(np.float)
accelZ = np.array(accelZ).astype(np.float)
gyroX = np.array(gyroX).astype(np.float)
gyroY = np.array(gyroY).astype(np.float)
gyroZ = np.array(gyroZ).astype(np.float)
magX = np.array(magX).astype(np.float)
magY = np.array(magY).astype(np.float)
magZ = np.array(magZ).astype(np.float)




cube = wf.Wireframe()
cube.addNodes([(300,300,300)])
cube.addNodes([(400,300,300)])
cube.addNodes([(300,400,300)])
cube.addNodes([(300,300,400)])
cube.addNodes([(200,300,300)])
cube.addNodes([(300,200,300)])
cube.addNodes([(300,300,200)])



cube.outputNodes()
cube.addEdges([(0,1)])
cube.addEdges([(0,2)])
cube.addEdges([(0,3)])
cube.addEdges([(0,4)])
cube.addEdges([(0,5)])
cube.addEdges([(0,6)])

viewer = wd.ProjectionViewer(800, 600)
viewer.addWireframe('cube', cube)
viewer.run()






```
